```python
import pandas as pd
from matplotlib import pyplot as plt
import re
import seaborn as sns
import numpy as np
```

```python
pd.set_option('display.max_rows',10)
data=pd.read_csv('data/breast-cancer-wisconsin2.data')
```

```python
data
```

```python
#number of benign tumors
pattern1=re.compile(r'(\d{5,7}),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),'+'2')
count1=0
with open('C:/Andrei/Python/flat_project/data/breast-cancer-wisconsin.data','r') as f:
    contents=f.read()
    matches=pattern1.finditer(contents)
    for match in matches:
        count1+=1
    print("Benign tumors: "+str(count1))
```

```python
#number of malignant tumors
pattern2=re.compile(r'(\d{5,7}),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),(\d+|\?),'+'4')
count2=0
with open('C:/Andrei/Python/flat_project/data/breast-cancer-wisconsin.data','r') as f:
    contents=f.read()
    matches=pattern2.finditer(contents)
    for match in matches:
        count2+=1
    print("Malignant tumors: "+str(count2))
```

```python
#bar chart benign vs malignant
plt.style.use('seaborn')
objects=("Benign","Malignant")
y_pos=np.arange(len(objects))
cases=[count1,count2]
plt.bar(y_pos,cases,align='center',alpha=0.5)
plt.xticks(y_pos,objects)
plt.ylabel("Cases")
plt.title("Tumors")
plt.show
```
